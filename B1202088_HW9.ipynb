```python
import cv2
import numpy as np
from urllib.request import urlopen
from matplotlib import pyplot as plt

def read_image_from_url(url):
    """Reads an image from a URL and returns it as a BGR numpy array."""
    response = urlopen(url)
    image = np.asarray(bytearray(response.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image

url = 'https://studiosol-a.akamaihd.net/uploadfile/letras/fotos/2/4/c/5/24c53722057633ab36e1fb7a23600790.jpg'

img_Lana_bgr = read_image_from_url(url)

print('Image Size',img_Lana_bgr.shape)

b, g, r = cv2.split(img_Lana_bgr)

img_Lana_rgb = cv2.merge([r, g, b])

fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 4))

def f_imshow(ax, mat, title):
    ax.imshow(mat, cmap='pink')
    ax.set_title(title)
    ax.axis('off')

f_imshow(axs[0], img_Lana_rgb, 'All Channels(RGB)')
f_imshow(axs[4], img_Lana_bgr, 'Original Channels(BGR)' )
f_imshow(axs[1], r, 'Red Channel')
f_imshow(axs[2], g, 'Green Channel')
f_imshow(axs[3], b, 'Blue Channel')

plt.tight_layout(w_pad=0.5)
plt.show()
```
